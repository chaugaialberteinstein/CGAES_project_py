### 2 Find the relationship between Job designation and mean Purchase amount

```
df.groupby('Job')['Purchase Price'].mean().sort_values(ascending = False).reset_index()
```

```
def decompose_categorical(x):
    if x=='Trade mark attorney':
        return 'Trade mark attorney'
    elif x=='Translator':
        return 'Translator'
    elif x=='Investment analyst':
        return 'Investment analyst'
    elif x=='Accountant, chartered management':
        return 'Accountant, chartered management'
    elif x=='Designer, industrial/product':
        return 'Designer, industrial/product'
    else:
        return 'Others'
    
df['Job']=df['Job'].apply(decompose_categorical)
```

```
mean_purchase_price= df.groupby('Job')['Purchase Price'].mean().sort_values().reset_index()
mean_purchase_price
```

```

# Create a bar plot
plt.figure(figsize=(18, 10))
mean_purchase_price.plot(kind='barh', x = 'Job', y = 'Purchase Price')
plt.xlabel('Job')
plt.ylabel('Mean Purchase Price')
plt.title('Mean Purchase Price by Job')
plt.xticks(rotation=80)
plt.show()
```

### 3 How does purchase value depend on the Internet Browser used and Job (Profession) of the purchaser?

```
df = pd.read_csv('Ecommerce_Purchases.csv')
df.head()
```

```
df3 = df[['Job','Purchase Price','Browser Info']]
df3
```

```
result = df3['Browser Info'].value_counts()
split_values = []
for i in range(len(result)):
    browser_info = result.index[i]
    matching_rows = df3['Browser Info'] == browser_info
    df3.loc[matching_rows, 'Browser'] = browser_info.split('/')[0]
```

```
df3['Browser'].value_counts()
```

```
df3 = df3.sort_values(by= 'Purchase Price', ascending = False)
df3
```

```
df3_2 = df3.groupby(['Job', 'Browser'])['Purchase Price'].sum().reset_index()

df3_2 = df3_2.pivot_table(index='Job', columns='Browser', values='Purchase Price', aggfunc='sum')
df3_2
```

```

# Plotting the DataFrame
df3_2[:10].plot(kind='bar', stacked=True, figsize=(10, 6))

# Adding labels and title
plt.xlabel('Job')
plt.ylabel('Purchase Price')
plt.title('Purchase Price by Job and Browser')

# Displaying the plot
plt.show()
```

### 4 What are the patterns, if any, on the purchases based on Location (State) and time of purchase (AM or PM)?

```
df['Address'].sample(5)
```

```
df['State'] = df['Address'].str.split(' ').str[-2]
df['State'].value_counts()
```

```
grouped_counts = df.groupby('State')['AM or PM'].value_counts()
grouped_counts
```

```
# Resetting the index to convert the Series to a DataFrame
grouped_counts = grouped_counts.reset_index(name='Count')

# Creating a pivot table
pivot_table = pd.pivot_table(grouped_counts, values='Count', index='State', columns='AM or PM', fill_value=0)

print(pivot_table)
```

```
pivot_table.sort_values(by='AM', ascending=False)[:10].plot.bar(color=['pink','purple'],legend = True)

# Set plot labels and title
plt.xlabel('State')
plt.ylabel('Count')
plt.title('AM Purchase Count by State')

# Display the plot
plt.show()
```

```
pivot_table.sort_values(by='PM', ascending=False)[:10].plot.bar(color=['yellow','maroon'],legend = True)

# Set plot labels and title
plt.xlabel('State')
plt.ylabel('Count')
plt.title('PM Purchase Count by State')

# Display the plot
plt.show()
```

### 6 What are top 5 Location(State) for purchases?

```
df['State'].value_counts().head()
```

```
def decompose(x):
    if x=='AP':
        return 'AP'
    elif x=='AA':
        return 'AA'
    elif x=='AE':
        return 'AE'
    elif x=='IN':
        return 'IN'
    elif x=='VI':
        return 'VI'
    else:
        return 'Others'
    
df['State']=df['State'].apply(decompose)
```

```
import pandas as pd
import matplotlib.pyplot as plt

# Create a pie chart of the 'State' column
df['State'].value_counts().plot.pie(autopct='%1.2f%%',  figsize=(5, 7))

# Set plot title
plt.title('State Distribution for Purchases')

# Display the plot
plt.show()
```
